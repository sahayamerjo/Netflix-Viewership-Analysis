8.	Write a program to plot using bar graph to view Total Viewership Hours by Release Season (2023)(tip: use the release month and convert them to winter,spring,summer,fall)

```python
def named_season(month):
    if month in ['Dec','Jan','Feb']:
        return 'Winter'
    elif month in ['Mar','Apr','May']:
        return 'Spring'
    elif month in ['Jun','Jul','Aug']:
        return 'Summer'
    elif month in ['Sep','Oct','Nov']:
        return 'Fall'
    else:
        return None
```

```python
Netflix['Season'] = Netflix['Released Month'].apply(named_season)
Netflix
```

```python
data_2023 = Netflix[Netflix['Released Year'] == 2023]
```

```python
group_season = data_2023.groupby('Season')["Hours Viewed"].sum()
plt.bar(group_season.index, group_season.values)
plt.xlabel('Released Season')
plt.ylabel("Total Viewership Hours")
plt.title("Total Viewership Hours by Released Season")
plt.show()
```

11.	Write a program to show the movies,series released on holidays such as new year,independence day, valentines day,christmas etc..

```python
Netflix['Release day'] = pd.to_datetime(Netflix['Release Date']).dt.day
Netflix
```

```python
def find_holidays(row):
    month = row['Released Month']
    day = row['Release day']
    
    if month == 'Jan' and day == 1:
        return 'New Year'
    elif month == 'Feb' and day == 14:
        return 'Valentine Day'
    elif month == 'Aug' and day == 15:
        return 'Independence Day'
    elif month == 'Dec' and day == 25:
        return 'Christmas'
    else:
        return None
```

```python
Netflix['Holiday'] = Netflix.apply(find_holidays, axis=1)
Netflix
```

```python
group_holiday = Netflix.groupby('Holiday')['Title'].size()
group_holiday
```

```python
plt.bar(group_holiday.index, group_holiday.values)
plt.xlabel('Special Holidays')
plt.ylabel('Total count of the Movies and Series')
plt.title('The Movies and Series Released on Holiday')
plt.show()
```
